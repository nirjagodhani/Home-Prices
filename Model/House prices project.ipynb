```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
```

```python
df1 = pd.read_csv("house_prices.csv")
df1.head()
```

```python
df1.shape
```

```python
df2= df1.drop(['date', 'street','statezip','country','waterfront','yr_built','yr_renovated','sqft_above'], axis='columns')
df2.head(20)
```

```python
df2.isnull().sum()
```

```python
df2.bedrooms.unique()
```

```python
df3=df2.copy()
df3['price_per_sqft']=df3['price']/df3['sqft_living']
df3.head()
```

```python
len(df3.city.unique())
```

```python
df3.city= df3.city.apply(lambda x: x.strip())

city_stats=df3.groupby('city')['city'].agg('count').sort_values(ascending=False)
city_stats
```

```python
len(city_stats[city_stats<=10])
```

```python
less_than_10 = city_stats[city_stats<=10]
df3.city= df3.city.apply(lambda x: 'other' if x in less_than_10 else x)
len(df3.city.unique())
```

```python
city_stats=df3.groupby('city')['city'].agg('count').sort_values(ascending=False)
city_stats
```

```python
df4=df3[~(df3.sqft_living/df3.bedrooms<300)]
df4.shape
```

```python
df4.price_per_sqft.describe()
```

```python
df4=df4.drop(df4[df4['price_per_sqft'] == 0].index)
df4.price_per_sqft.describe()
```

```python
df4=df4.drop(df4[df4['bedrooms'] == 0].index)
df4.bedrooms.describe()
```

```python
df4=df4.drop(df4[df4['price_per_sqft'] >= 2000].index)
df4.price_per_sqft.describe()
```

```python
dummies = pd.get_dummies(df4.city)
dummies.head()
```

```python
df5 = pd.concat([df4,dummies.drop('other',axis = 'columns')],axis='columns')
df5.head()
```

```python
df6 = df5.drop('city',axis='columns')
df6.head(2)
```

```python
df6.shape
```

```python
df7= df6.drop(['price_per_sqft'], axis='columns')
df7.head()
```

```python
X= df7.drop(['price',],axis='columns')
X.head(3)
```

```python
y= df5.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=5)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
import pickle
with open('Home_prices_model.pickle','wb') as f:
    pickle.dump(lr,f)
```
